Importing the required libraries for data processing

```python
import requests
import json
import pandas as pd
```

Fetching the data from REST API url and parsing the data into json followed by conversion to data frame

```python
url = "https://restcountries.com/v3.1/all"
params = {"fields": "name,population,region,area,languages,borders,unMember,independent,subregion,continents"}

response = requests.get(url, params)
count_data = response.json()
df = pd.DataFrame(count_data)
df['common_name'] = df['name'].apply(lambda x: x['common'])
df['official_name'] = df['name'].apply(lambda x: x['official'])
df['languages'] = df['languages'].apply(lambda x: ', '.join(sorted(x.values())))
df['continents'] = df['continents'].str.join(', ')
df['borders'] = df['borders'].str.join(', ')

df
```

Re-arranging the columns for better data corelation and analysis

```python
df.drop(columns = ['name'], inplace = True)
df = df[['common_name', 'continents', 'region', 'subregion','independent','unMember','languages','area','population','borders']]
df
```

Total Number of countries per continent

```python
con_df = df.groupby('continents')['common_name'].count().rename('Total_countries').sort_values( ascending=False).reset_index()
con_df
```

Total Number of countries per region

```python
Tot_df = df.groupby('region')['common_name'].count().rename('Total_countries').sort_values(ascending=False).reset_index()
Tot_df
```

Number of independent countries per region

```python
ind_df = df.groupby('region')['independent'].sum().rename('Independent_countries').sort_values(ascending=False).reset_index()
ind_df
```

Total number of non-independent countries per region

```python
df_merged = pd.merge(Tot_df, ind_df, on='region')
df_merged['Non_independent_countries'] = df_merged['Total_countries'] - df_merged['Independent_countries']
df_merged.sort_values(by='Non_independent_countries', ascending=False)
```

Number of UN members per continent

```python
un_df = df.groupby('continents')['unMember'].sum().rename('Total_UN_members').sort_values(ascending=False).reset_index()
un_df
```

```python
df_merged_1 = pd.merge(con_df, un_df, on='continents')
df_merged_1['Non_UN_members'] = df_merged_1['Total_countries'] - df_merged_1['Total_UN_members']
df_merged_1.sort_values(by='Non_UN_members', ascending=False)
```

Number of UN members per region

```python
un_df_1 = df.groupby('region')['unMember'].sum().rename('Tot_reg_UN_members').sort_values(ascending=False).reset_index()
un_df_1
```

```python
df_merged_2 = pd.merge(Tot_df, un_df_1, on='region')
df_merged_2['Non_reg_UN_members'] = df_merged_2['Total_countries'] - df_merged_2['Tot_reg_UN_members']
df_merged_2.sort_values(by='Non_reg_UN_members', ascending=False)
```
