Importing the required libraries

```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

Loading data/trends_analysed.csv into a DataFrame

```python
df = pd.read_csv('/content/data/trends_analysed.csv')
df = df.drop(df.columns[0], axis=1)
df
```

Creating a folder called outputs/

```python
from pathlib import Path
Path("output").mkdir(exist_ok=True)
```

Creating a horizontal bar chart showing the top 10 stories by score

```python
xx = df[['score','title']].sort_values(by='score',ascending=False).head(10)
xx
```

```python
xx.plot.barh(x='title', y='score', figsize=(20,10))
plt.title("Top 10 Stories by Score")
plt.xlabel("Score")
plt.ylabel("Title")
plt.savefig('output/chart1_top_stories.png')
#
```

Create a bar chart showing how many stories came from each category

```python
y = df.groupby('category')['post_id'].count().reset_index()
colors = ['red', 'blue', 'green', 'orange', 'gold']
y.plot(kind='bar', x='category', y='post_id', figsize=(10,10), color = colors)
plt.title("Number of Stories per Category")
plt.xlabel("Category")
plt.ylabel("Number of Stories")
plt.savefig('output/chart2_categories.png')
#
```

Create a scatter plot with score on the x-axis and num_comments on the y-axis

Colour the dots differently for popular vs non-popular stories (use the is_popular column)

```python
sns.scatterplot(data=df, x='score', y='num_comments', hue = 'is_popular')
plt.title("Score vs Number of Comments")
plt.xlabel("Score")
plt.ylabel("Number of Comments")
plt.savefig('output/chart3_scatter.png')
#
```
