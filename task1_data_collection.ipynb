Importing the required libraries

```python
import requests
import json
import pandas as pd
```

Fetching the ID as a json file

```python
api_url = 'https://hacker-news.firebaseio.com/v0/topstories.json'
api_url_2 = 'https://hacker-news.firebaseio.com/v0/item/{id}.json'
stories = {}

data = requests.get(url = api_url, headers = {"User-Agent": "TrendPulse/1.0"})
ids = data.json()

```

Fetching individual story per ID and defining categories as per their title

```python
stories = {}
Technology = ['AI', 'software', 'tech', 'code', 'computer', 'data', 'cloud', 'API', 'GPU', 'LLM']
Worldnews = ['war', 'government', 'country', 'president', 'election', 'climate', 'attack', 'global']
Sports = ['NFL', 'NBA', 'FIFA', 'sport', 'game', 'team', 'player', 'league', 'championship']
Science = ['research', 'study', 'space', 'physics', 'biology', 'discovery', 'NASA', 'genome']
Entertainment = ['movie', 'film', 'music', 'Netflix', 'game', 'book', 'show', 'award', 'streaming']
for id in ids:
  response = requests.get(url = api_url_2.format(id = id), headers = {"User-Agent": "TrendPulse/1.0"})
  if response.status_code == 200:
    stories[id] = (requests.get(url = api_url_2.format(id = id), headers = {"User-Agent": "TrendPulse/1.0"})).json()
  else:
    print(response.status_code)
    continue
  #print(response.status_code)
#print(stories)

category = {}
collected_stories = {}
for story in stories.values():
  for y in Sports:
    if y in story['title'].split():
      category[story['id']] = 'sports'
  for z in Science:
    if z in story['title'].split():
      category[story['id']] = 'science'
  for w in Entertainment:
    if w in story['title'].split():
      category[story['id']] = 'entertainment'
  for t in Technology:
    if t  in story['title'].split():
      category[story['id']] = 'technology'
  for x in Worldnews:
    if x in story['title'].split():
      category[story['id']] = 'worldnews'

category
```

Converting json file into data frames so that i can perform multiple operations on them

```python
category_df = pd.DataFrame.from_dict(category, orient='index')
category_df.rename(columns={0: 'category'}, inplace=True)
category = category_df.reset_index()
category.sort_values(by=['index'], inplace=True)
category.rename(columns={'index': 'post_id'}, inplace=True)
category

```

```python
stories_df = pd.DataFrame(stories).transpose().reset_index()
stories_df.head()
stories_df.transpose()
stories_df.reset_index().head(50)
stories_df.sort_values(by=['index'], inplace=True)
stories_df.rename(columns={'index': 'post_id'}, inplace=True)
stories_df
```

```python
un_df = pd.merge(category, stories_df, on='post_id', how='outer')
un_df.rename(columns = {'descendants':'num_comments', 'by': 'author'}, inplace=True)
un = un_df.drop(columns=['id', 'kids', 'type', 'url', 'text', 'time'])
un['collected_at'] = pd.Timestamp.now()
Tot_Stories = un['post_id'].count()
un




```

Making a new folder called data

```python
from pathlib import Path
Path("data").mkdir(exist_ok=True)

```

Converting the data frame to json file and saving in data folder

```python
un.to_json("data/trends_20260409.json")
saved_path = "data/trends_20260409.json"
print(f"Collected {Tot_Stories} stories. Saved to {saved_path}")
```
